# Financial Analysis Report

## Introduction
This notebook analyzes the financial performance of Microsoft, Tesla, and Apple over the last three fiscal years. The focus is on:
- Revenue trends
- Net income changes
- Assets vs. liabilities
- Cash flow trends

```python

excel_file_path = '/Users/srikanthpusthem/Desktop/Projects/bcg-job-simulation/Financial_Data_10K.xlsx'

csv_file_path = '/Users/srikanthpusthem/Desktop/Projects/bcg-job-simulation/Financial_Data_10K.csv'
df = pd.read_excel(excel_file_path)

df.to_csv(csv_file_path, index=False)
print(df.head())
```

## Key Insights

### Revenue Growth Trends
1. Apple showed stable revenue trends with no significant year-over-year growth or decline.
2. Microsoft experienced an average negative revenue growth of -10.62%.
3. Tesla's revenue growth was the most volatile with -16.99% on average.

### Net Income Trends
- Microsoft had declining net income for two consecutive years (2023 and 2022).
- Tesla and Apple showed no declines in net income during the period.

### Assets vs. Liabilities
- Apple had the highest total assets ($352.8 billion) and liabilities ($283.3 billion).
- Microsoft maintained a balanced financial structure.
- Tesla had the lowest assets and liabilities, reflecting its smaller scale.

### Cash Flow Trends
- Apple led with $122.2 billion average cash flow, followed by Microsoft with $91 billion.
- Tesla had $13.83 billion, reflecting its evolving business model.


```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue (B)'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income (B)'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets (B)'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities (B)'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby(['Company'])['Cash Flow from Operating Activities (B)'].pct_change() * 100
```

```python
summary = df.groupby('Company')[['Revenue Growth (%)', 'Net Income Growth (%)']].mean()
print(summary)
```

## Visualizations
The plot below compares revenue trends across companies over the last three fiscal years.


```python
import matplotlib.pyplot as plt

# Define the financial metric you want to compare (e.g., Total Revenue)
financial_metric = 'Total Revenue (B)'

# Create a new figure
plt.figure(figsize=(12, 8))

# Loop through each company to plot their data
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(
        company_data['Fiscal Year Ended'], 
        company_data[financial_metric], 
        marker='o', 
        label=company
    )

# Add titles and labels
plt.title(f"Comparison of {financial_metric} Across Companies", fontsize=16)
plt.xlabel("Fiscal Year", fontsize=12)
plt.ylabel(f"{financial_metric} (Billion $)", fontsize=12)
plt.legend(title="Company", fontsize=10)
plt.grid(True)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
